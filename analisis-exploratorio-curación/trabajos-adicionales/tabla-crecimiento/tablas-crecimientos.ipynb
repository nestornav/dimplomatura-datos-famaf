```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# Importaci√≥n de los dataset
weight_age = pd.read_csv('./dataset/wtageinf.csv')
length_age = pd.read_csv('./dataset/lenageinf.csv')
```

```python
# Obtengo los datos relacionados al peso por sexo, primero maculino y luego femenino.
boys = weight_age.loc[weight_age['Sex'] == 1]

plt.title('Boys weight until 36 months')
plt.ylabel('Weight (Kg)')
plt.xlabel('Month')

plt.plot(boys['P3'])
plt.plot(boys['P5'])
plt.plot(boys['P10'])
plt.plot(boys['P25'])
plt.plot(boys['P50'])
plt.plot(boys['P75'])
plt.plot(boys['P90'])
plt.plot(boys['P95'])
plt.plot(boys['P97'])

plt.show()
boys.head()
```

```python
girls = weight_age.loc[weight_age['Sex'] == 2]

plt.title('Girls weight until 36 months')
plt.ylabel('Weight (Kg)')
plt.xlabel('Month')

plt.plot(girls['P3'])
plt.plot(girls['P5'])
plt.plot(girls['P10'])
plt.plot(girls['P25'])
plt.plot(girls['P50'])
plt.plot(girls['P75'])
plt.plot(girls['P90'])
plt.plot(girls['P95'])
plt.plot(girls['P97'])

girls.head()
```

```python
# Dada las primeras pruebas los resultados que se obten√≠an no eran correctos.
# Por lo que se verifica el tipo de valor de las columnas. Encontramos que son del tipo string :(.
dtypeCount =[length_age.iloc[:,i].apply(type).value_counts() for i in range(length_age.shape[1])]

dtypeCount
```

```python
# Vamos a convertir los valores necesarios en num√©ricos.

length_age['sex'] = pd.to_numeric(length_age['Sex'], errors='coerce')
length_age['pub3'] = pd.to_numeric(length_age['Pub3'], errors='coerce')
length_age['pub5'] = pd.to_numeric(length_age['Pub5'], errors='coerce')
length_age['pub10'] = pd.to_numeric(length_age['Pub10'], errors='coerce')
length_age['pub25'] = pd.to_numeric(length_age['Pub25'], errors='coerce')
length_age['pub50'] = pd.to_numeric(length_age['Pub50'], errors='coerce')
length_age['pub75'] = pd.to_numeric(length_age['Pub75'], errors='coerce')
length_age['pub90'] = pd.to_numeric(length_age['Pub90'], errors='coerce')
length_age['pub95'] = pd.to_numeric(length_age['Pub95'], errors='coerce')
length_age['pub97'] = pd.to_numeric(length_age['Pub97'], errors='coerce')
```

```python
boys_len = length_age.loc[length_age['sex'] == 1]

plt.title('Boys length until 36 months')
plt.ylabel('Length (cm)')
plt.xlabel('Month')

plt.plot(boys_len['pub3'])
plt.plot(boys_len['pub5'])
plt.plot(boys_len['pub10'])
plt.plot(boys_len['pub25'])
plt.plot(boys_len['pub50'])
plt.plot(boys_len['pub75'])
plt.plot(boys_len['pub90'])
plt.plot(boys_len['pub95'])
plt.plot(boys_len['pub97'])


plt.show()
```

```python
girls_len = length_age.loc[length_age['sex'] == 2]

plt.title('Girls length until 36 months')
plt.ylabel('Length (cm)')
plt.xlabel('Month')

plt.plot(girls_len['pub3'])
plt.plot(girls_len['pub5'])
plt.plot(girls_len['pub10'])
plt.plot(girls_len['pub25'])
plt.plot(girls_len['pub50'])
plt.plot(girls_len['pub75'])
plt.plot(girls_len['pub90'])
plt.plot(girls_len['pub95'])
plt.plot(girls_len['pub97'])

plt.show()
```
